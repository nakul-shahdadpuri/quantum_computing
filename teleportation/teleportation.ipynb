```python
from qiskit import *
```

```python
circuit = QuantumCircuit(3,3)
```

```python
circuit.h(0)
```

```python
circuit.barrier()
```

```python
circuit.h(1)
circuit.cx(1,2)
```

```python
%matplotlib inline
circuit.draw(output = "mpl")
```

```python
circuit.cx(0,1)
circuit.h(0)
```

```python
%matplotlib inline
circuit.draw(output = "mpl")
```

```python
circuit.barrier()
circuit.measure([0,1],[0,1])
circuit.draw(output = "mpl")
```

```python
circuit.barrier()
circuit.cx(1,2)
circuit.cz(0,2)
circuit.draw(output = "mpl")
```

```python
circuit.measure(2,2)
simulator = Aer.get_backend('qasm_simulator')
results = execute(circuit,backend = simulator,shots = 1024).result()
counts = results.get_counts()
from qiskit.tools.visualization import plot_histogram
plot_histogram(counts)
```
